# **What is Object-Oriented Programming (OOP)?**

OOP is a programming model designed to organize software design around data (objects) rather than functions and logic. Objects are instances of classes, which can hold both data and methods.

# 🔹 Core Concepts of OOP

# **1. Class**
A blueprint or template for creating objects.

Defines attributes (data members) and methods (functions) that the object will have.

Example (in Python):

        self.model = model

    print("The car is driving")

# **2. Object**
An instance of a class.

Represents a specific example of a class with actual values.

# **3. Encapsulation**
Bundling of data (attributes) and methods that operate on that data into a single unit (class).

Protects the internal state of an object from unintended modification.

Achieved through access modifiers (public, private, protected).


class Person:
    def __init__(self, name):
        self.__name = name  # private attribute

    def get_name(self):
        return self.__name


# **4. Abstraction**
Hides complex implementation details and shows only the necessary features.

Helps in reducing programming complexity.

Achieved through abstract classes or interfaces (in some languages).

from abc import ABC, abstractmethod

class Animal(ABC):  # Abstract class

    @abstractmethod
    def sound(self):  # Abstract method
        pass

class Dog(Animal):

    def sound(self):
        return "Bark"

class Cat(Animal):

    def sound(self):
        return "Meow"

# Usage
animals = [Dog(), Cat()]
for a in animals:
    print(a.sound())

# **6. Polymorphism**
Ability of different objects to respond to the same function call in different ways.

Example: Method Overriding.

# **🔹 Common OOP Terminologies**
Constructor: A special method used to initialize objects.

Destructor: A method called when an object is destroyed.

Method Overloading: Same method name, different parameters (compile-time polymorphism).

Method Overriding: Subclass provides a specific implementation of a method already defined in the parent class (run-time polymorphism).

# **2- Why is OOP important in Python and AI?**

Object-Oriented Programming (OOP) is important in Python and AI development for several practical and structural reasons. Here’s a breakdown of why it matters:

# 💡 Real-World Example
In PyTorch:

import torch.nn as nn

class NeuralNet(nn.Module):
    def __init__(self):
        super(NeuralNet, self).__init__()
        self.layer1 = nn.Linear(10, 50)
        self.relu = nn.ReLU()
        self.output = nn.Linear(50, 1)
    
    def forward(self, x):
        x = self.relu(self.layer1(x))
        return self.output(x)

This OOP structure enables clear abstraction of layers and logic, making it scalable and clean.

# **1. Class**

A class is a blueprint or template for creating objects. It defines the structure and behavior (data and functions) that the objects created from the class will have.

Example:


class Dog:

    def __init__(self, name, breed):
        self.name = name
        self.breed = breed
    
    def bark(self):
        print(f"{self.name} says woof!")

Dog is a class.

# 2. Object

An object is an actual instance of a class. It represents a specific entity created based on the class blueprint, with its own values for attributes.

Example:

my_dog = Dog("Buddy", "Golden Retriever")

my_dog is an object of the Dog class.

# 3. Instance
An instance is essentially another word for an object. When you create an object from a class, that object is an instance of the class.

Example:

isinstance(my_dog, Dog)  # True

my_dog is an instance of the class Dog.

# 4. Method

A method is a function defined inside a class that operates on instances of that class. It usually takes self as the first parameter to refer to the specific object calling the method.

Example:

def bark(self):

    print(f"{self.name} says woof!")

bark is a method of the class Dog.

# 5. Attribute

An attribute is a variable that belongs to an object. It stores data about the object and is defined typically in the __init__ method.

Example:

self.name = name

self.breed = breed

name and breed are attributes of the Dog object.

❖ Task 1.1: Create a Car Class Create a class named Car with the following: • Attributes: brand, color • Method: display_info() to print brand and color Create two objects of Car with different values and display their info. python

```python
# Define the Car class
class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def display_info(self):
        print(f"Brand: {self.brand}, Color: {self.color}")

# Create two Car objects with different values
car1 = Car("Toyota", "Red")
car2 = Car("BMW", "Black")

# Display their information
car1.display_info()
car2.display_info()
```

❖ Task 1.2: Create a Student Class Your Own Practice: • Class: Student • Attributes: name, age, grade • Method: get_details() to print student info

```python
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def get_details(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")
        print(f"Grade: {self.grade}")

# Example usage:
student1 = Student("Alice", 15, "10th Grade")
student1.get_details()
```
