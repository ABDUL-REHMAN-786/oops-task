# **1. What is Encapsulation in OOP?**

# üîí Definition:
Encapsulation is the concept of wrapping data (variables) and code (methods) together into a single unit (class). It also involves restricting direct access to some of the object‚Äôs components, which is a way of data hiding.

# üß† Purpose:
Prevent unintended interference and misuse of data.

Improve modularity and maintainability of the code.

# ‚úÖ Key Idea:
Only expose what‚Äôs necessary and hide internal object details from the outside world.

# üßæ Example:

class BankAccount:
    def __init__(self, balance):
        self.__balance = balance  # private variable

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
print(account.get_balance())  # 1500
print(account.__balance)  ‚ùå Raises AttributeError (data is encapsulated)


# 2. What are Access Modifiers (public, private, protected) in Python?
Python doesn‚Äôt have true access modifiers like Java/C++, but it follows naming conventions:

# ‚úÖ Access Modifiers in Python (List Format):

# 1- Public (variable)

Convention: variable

Accessible From: Anywhere

Description:

This is the default access level in Python.

Can be accessed from outside or inside the class.

No restrictions on access.

# 2- Protected (_variable)

Convention: _variable (single underscore prefix)

Accessible From: Within the class and its subclasses

Description:

Indicates that the variable is for internal use only.

It's just a convention, not enforced by Python.

Can still be accessed from outside, but it's discouraged.


# 3- Private (__variable)

Convention: __variable (double underscore prefix)

Accessible From: Within the class only

Description:

Python uses name mangling: __var becomes _ClassName__var.

This makes it harder (but not impossible) to access from outside.

Provides stronger encapsulation compared to protected.



# 3. Why and when should we hide data in real-world projects?

**‚úÖ Why Should We Hide Data?**
**Maintainability**

Hiding data makes code easier to update or refactor. If internal details are hidden, changes can be made without affecting other parts of the system.

**Security**

Sensitive information (like passwords, API keys, or internal logic) should be inaccessible from outside the class/module to prevent misuse or attacks.

**Prevent Misuse**

If internal data or behavior is exposed, external code might depend on it in unintended ways, causing bugs or instability.

**Encapsulation**

Hiding data supports object-oriented principles, encapsulating the state and behavior of objects, allowing controlled access through public methods (getters/setters).

**Improved Debugging**

When data is accessed only through specific methods, it‚Äôs easier to track bugs and add validation or logging.

**Flexibility and Modularity**

The implementation details can be changed without impacting external code as long as the interface remains consistent.

**üïí When Should We Hide Data?**
**During Class Design (Object-Oriented Programming)**

From the beginning, when designing classes, internal variables (attributes) should be declared private or protected.

**When Handling Sensitive Information**

Always hide sensitive data like user credentials, tokens, and internal algorithms.

**During API Design**

When exposing functionality to other systems or developers, keep the internal logic hidden and only expose what is necessary.

**In Large-Scale Systems**

As systems grow, clear boundaries and hidden data ensure that modules/components don‚Äôt interfere with each other.

**Before Public Release**

Ensure internal workings are abstracted before releasing software or libraries to avoid breaking user code in future updates.

# **4. What Are Getter and Setter Methods? Why Are They Important?**
**‚úÖ What Are Getter and Setter Methods?**

Getter and setter methods are special functions used to access (get) and modify (set) the private or protected fields (attributes) of a class in object-oriented programming (OOP).

**üîπ Getter Method**

Retrieves the value of a private variable.

Typically named with the prefix get.

def get_balance(self):
    return self.__balance

**üîπ Setter Method**

Updates or modifies the value of a private variable.

Usually includes validation.

Named with the prefix set.

def set_balance(self, new_balance):
    if new_balance >= 0:
        self.__balance = new_balance

**‚úÖ Why Are Getter and Setter Methods Important?**

**Encapsulation**

Protects the internal state of an object by controlling how its data is accessed or changed.

Only exposes what‚Äôs necessary.

**Data Validation**

Setters allow you to validate inputs before updating a variable, preventing invalid states.

Example: Preventing a bank account from having a negative balance.

**Security**

Sensitive data can be read-only or write-only as needed.

Prevents unauthorized access or modification.

**Flexibility**

You can change the internal implementation later (e.g., change how a value is calculated or stored) without affecting code that uses the getter or setter.

**Debugging and Logging**

Easy to add logging, debugging, or audit functionality within getters/setters.

**Compatibility**

Provides a consistent interface even if the internal representation changes.

class Student:
    def __init__(self, name):
        self.__name = name

    def get_name(self):
        return self.__name

    def set_name(self, new_name):
        if new_name != "":
            self.__name = new_name

**üß† Summary:**

**Method:** Getter
**Purpose:** Access private data
**Example:** get_name()

**Method:** Setter
**Purpose:** Modify private data
**Example:** set_name("Alice")

Using getters and setters is a best practice in OOP to write secure, maintainable, and modular code.

```python
class BankAccount:
    def __init__(self, account_holder, initial_balance):
        self.account_holder = account_holder
        if initial_balance >= 0:
            self.__balance = initial_balance
        else:
            self.__balance = 0
            print("Initial balance cannot be negative. Setting balance to 0.")

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Depositing {amount}...")
        else:
            print("Deposit amount must be positive!")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawing {amount}...")
        else:
            print("Insufficient balance!")

    def get_balance(self):
        return self.__balance

    def set_balance(self, amount):
        if amount >= 0:
            self.__balance = amount
            print(f"Balance set to {amount}.")
        else:
            print("Invalid balance amount! Balance not updated.")

    def account_summary(self):
        print(f"Account Holder: {self.account_holder}")
        print(f"Current Balance: {self.__balance}")


# ‚úÖ Creating an object with sample data
account1 = BankAccount("Ali", 0)

# ‚ùå Try accessing __balance directly (should fail)
try:
    print(account1.__balance)
except AttributeError as e:
    print("Direct access to __balance failed:", e)

# ‚úÖ Using methods to interact with balance
account1.deposit(500)
account1.withdraw(200)
print("Current balance:", account1.get_balance())

# ‚úÖ Use setter (e.g., reset balance)
account1.set_balance(1000)
print("New balance:", account1.get_balance())

# ‚úÖ Print account summary
account1.account_summary()

print("\n--- Creating More Accounts ---\n")

# ‚úÖ Bonus: Creating more accounts
account2 = BankAccount("Sara", 300)
account3 = BankAccount("John", 1000)

account2.deposit(-50)  # Should fail
account2.withdraw(100)
account2.account_summary()

account3.deposit(500)
account3.withdraw(2000)  # Should fail due to insufficient funds
account3.account_summary()
```
